```python

print(" Rules of the Game:This game will consist of 10 rounds.Player will be declared winner who wins maximum no of rounds")
i=1
w=0
w1=0
ch1="y"
while ch1=="y":
    while i<=10:
        i=i+1
        ch=input("Enter your choice S for Snake/W for Water/G for Gun")
        if(ch=="S"):
            print("Your choice:Snake")
        elif(ch=="W"):
            print("Your choice:Water")
        elif(ch=="G"):
            print("Your choice:Gun")
        import random
        list=["Snake","Water","Gun"]
        choice=random.choice(list)
        print("Computer choice:",choice)
        if(ch=="S" and choice=="Snake"):
            print("No result")
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="S" and choice=="Water"):
            print("Congrats you won this round!")
            w+=1
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="S" and choice=="Gun"):
            print("Sorry you lose this round!")
            w1+=1
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="W" and choice=="Water"):
            print("No result")
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="W" and choice=="Snake"):
            print("Sorry you lose this round!")
            w1+=1
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="W" and choice=="Gun"):
            print("Congrats you won this round!")
            w=w+1
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="G" and choice=="Water"):
            print("Sorry you lose this round!")
            w1+=1
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="G" and choice=="Snake"):
            print("Congrats you won this round!")
            w+=1
            print("Your score:",w, "Computer score:",w1)
        elif(ch=="G" and choice=="Gun"):
            print("No result")
            print("Your score:",w, "Computer score:",w1)
    if w>w1:
     print("Congrats you won the game")
    elif w==w1:
     print("Game draw")
    else:
     print("Sorry you lose the game")
     print("Your score",w)
     print("Computer score",w1)
     ch1=input("Type y to play again")

print("Bbye see you later")




    
    
    
    
    
    
    
    
    
    
    




```
